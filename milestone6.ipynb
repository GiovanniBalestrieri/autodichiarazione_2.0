* Classi e funzioni per l'intersezione tra due segmenti. Thanks to https://bryceboe.com/2006/10/23/line-segment-intersection-algorithm/

```python
import numpy as np
import pylab as pl
from matplotlib import collections  as mc

class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

def ccw(A,B,C):
    return (C.y-A.y) * (B.x-A.x) > (B.y-A.y) * (C.x-A.x)

# Return true if line segments AB and CD intersect
def intersect(A,B,C,D):
    return ccw(A,C,D) != ccw(B,C,D) and ccw(A,B,C) != ccw(A,B,D) 

def plot_intersect(segm_a, segm_b):
    lines = [[segm_a[0], segm_a[1]], [segm_b[0], segm_b[1]]]
    c = np.array([(1, 0, 0, 1), (0, 1, 0, 1)])

    lc = mc.LineCollection(lines, colors=c, linewidths=2)
    fig, ax = pl.subplots()
    ax.add_collection(lc)
    ax.autoscale()
    ax.margins(0.1)
```
